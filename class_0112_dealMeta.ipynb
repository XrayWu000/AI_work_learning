```python
import glob
import json
import pandas as pd
import datetime as dt

def change_datetype(s):
    d = dt.datetime.strptime(s, "%a %b %d %H:%M:%S %Y")
    s2 = d.strftime("%Y/%m/%d %H:%M:%S")
    return s2

table = []
for fp in glob.glob("./PTT/*/meta.json"):
    f = open(fp, "r", encoding="utf-8")
    content = f.read()
    f.close()
    row = json.loads(content)
    table.append(row)
df = pd.DataFrame(table)
df["category"] = df["title"].str.split("[", expand=True)[1].str.split("]").str[0]
df["title"] = df["title"].str.split("]").str[1]
idx = df.columns.get_loc("title")
df.insert(idx, "category", df.pop("category"))
df["datetime(format)"] = df["datetime"].apply(change_datetype)
idx = df.columns.get_loc("datetime")
df.insert(idx, "datetime(format)", df.pop("datetime(format)"))
df
```
